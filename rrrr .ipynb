```python
import matplotlib.pyplot as plt
import numpy as np

age_1 = np.random.randint(15,35, (100)) 
age_2 = np.random.randint(15,35, (100))

bins_data = [15, 20, 25, 30, 35]
plt.hist([age_1, age_2], color=['red', 'yellow'], orientation='vertical', rwidth=0.8, bins=bins_data, histtype='bar', label=['Age 1', 'Age 2'], align='mid')
plt.legend()
# Set custom x-axis ticks at the edges of the bins
plt.xticks(bins_data)
plt.show()

```
