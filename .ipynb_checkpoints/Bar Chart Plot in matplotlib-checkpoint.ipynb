# Bar Chart Plot

```python
import matplotlib.pyplot as plt 
x = [ 'python' , 'c' , 'c++', 'java'] ; y = [80,70,75,83]
plt.xlabel('Programming Language', fontsize = 15) ; plt.ylabel( 'No. of Users', fontsize = 15) ; plt.title( ' Users vs Language Bar Chart', fontsize = 15)
color_name = [ 'r', 'y' , 'b' , 'g' ]
plt.bar(x,y, width = 0.4, color = color_name)
plt.show()
```

```python
x1 = [ 'python' , 'c' , 'c++', 'java'] ; y1 = [80,70,75,83]
plt.xlabel('Programming Language', fontsize = 15) ; plt.ylabel( 'No. of Users', fontsize = 15) ; plt.title( ' Users vs Language Bar Chart', fontsize = 15)
color_name1 = [ 'r', 'y' , 'b' , 'g' ]
plt.bar(x1,y1, width = 0.4, color = color_name1, align = 'edge' ) # x-axis value will be start from the starting edge of bar
plt.show()
```
